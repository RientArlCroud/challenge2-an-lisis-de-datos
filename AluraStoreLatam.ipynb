### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## Gráficos


### Este gráfico va a representar los ingresos de las tiendas por cada año

1. tienda 1:


```python
tienda['Fecha de Compra'] = pd.to_datetime(tienda['Fecha de Compra'], dayfirst=True, errors='coerce')
tienda['Año'] = tienda['Fecha de Compra'].dt.year
ingresos_ct_año = tienda.groupby('Año')[['Precio']].sum()
print(ingresos_ct_año)
```

2. tienda 2:


```python
tienda2['Fecha de Compra'] = pd.to_datetime(tienda2['Fecha de Compra'], dayfirst=True, errors='coerce')
tienda2['Año'] = tienda2['Fecha de Compra'].dt.year
ingresos_ct_año2 = tienda2.groupby('Año')[['Precio']].sum()
print(ingresos_ct_año2)
```

3. tienda 3:


```python
tienda3['Fecha de Compra'] = pd.to_datetime(tienda3['Fecha de Compra'], dayfirst=True, errors='coerce')
tienda3['Año'] = tienda3['Fecha de Compra'].dt.year
ingresos_ct_año3 = tienda3.groupby('Año')[['Precio']].sum()
print(ingresos_ct_año3)
```

4. tienda 4:


```python
tienda4['Fecha de Compra'] = pd.to_datetime(tienda4['Fecha de Compra'], dayfirst=True, errors='coerce')
tienda4['Año'] = tienda4['Fecha de Compra'].dt.year
ingresos_ct_año4 = tienda4.groupby('Año')[['Precio']].sum()
print(ingresos_ct_año4)
```

## Vamos a unir todas las tiendas en un solo DF
### para eso primero limpiamos los datos

```python
ingresos_ct_año = ingresos_ct_año.rename(columns={'Precio':'Tienda 1'})
ingresos_ct_año2 = ingresos_ct_año2.rename(columns={'Precio':'Tienda 2'})
ingresos_ct_año3 = ingresos_ct_año3.rename(columns={'Precio':'Tienda 3'})
ingresos_ct_año4 = ingresos_ct_año4.rename(columns={'Precio':'Tienda 4'})
```

```python
ventas_tiendas = pd.concat([ingresos_ct_año, ingresos_ct_año2, ingresos_ct_año3, ingresos_ct_año4], axis=1)
ventas_tiendas
```
