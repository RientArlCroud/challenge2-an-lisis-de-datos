### Este gráfico va a representar los ingresos de las tiendas por cada año

1. tienda 1:


```python
tienda['Fecha de Compra'] = pd.to_datetime(tienda['Fecha de Compra'], dayfirst=True, errors='coerce')
tienda['Año'] = tienda['Fecha de Compra'].dt.year
ingresos_ct_año = tienda.groupby('Año')[['Precio']].sum()
print(ingresos_ct_año)
```

3. tienda 3:


```python
tienda3['Fecha de Compra'] = pd.to_datetime(tienda3['Fecha de Compra'], dayfirst=True, errors='coerce')
tienda3['Año'] = tienda3['Fecha de Compra'].dt.year
ingresos_ct_año3 = tienda3.groupby('Año')[['Precio']].sum()
print(ingresos_ct_año3)
```

4. tienda 4:


```python
tienda4['Fecha de Compra'] = pd.to_datetime(tienda4['Fecha de Compra'], dayfirst=True, errors='coerce')
tienda4['Año'] = tienda4['Fecha de Compra'].dt.year
ingresos_ct_año4 = tienda4.groupby('Año')[['Precio']].sum()
print(ingresos_ct_año4)
```

## Vamos a unir todas las tiendas en un solo DF
### para eso primero limpiamos los datos

```python
ingresos_ct_año = ingresos_ct_año.rename(columns={'Precio':'Tienda 1'})
ingresos_ct_año2 = ingresos_ct_año2.rename(columns={'Precio':'Tienda 2'})
ingresos_ct_año3 = ingresos_ct_año3.rename(columns={'Precio':'Tienda 3'})
ingresos_ct_año4 = ingresos_ct_año4.rename(columns={'Precio':'Tienda 4'})
```

```python
ventas_tiendas = pd.concat([ingresos_ct_año, ingresos_ct_año2, ingresos_ct_año3, ingresos_ct_año4], axis=1)
ventas_tiendas
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

for tienda in ventas_tiendas.columns:
  plt.plot(
      ventas_tiendas.index,
      ventas_tiendas[tienda],
      marker='o',
      linewidth=2,
      label=tienda
  )

plt.title('Ingresos por tienda a lo largo del tiempo')
plt.xlabel('Año')
plt.ylabel('Ingresos totales (M)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
ventas_tiendas = ventas_tiendas.drop(2023)
```

```python
ventas_millones = ventas_tiendas / 1000000
ventas_millones
```

```python
plt.figure(figsize=(12,6))

for tienda in ventas_tiendas.columns:
  plt.plot(
      ventas_tiendas.index,
      ventas_millones[tienda],
      marker='o',
      linewidth=2,
      label=tienda
  )



plt.title('Ingresos por tienda a lo largo del tiempo')
plt.xlabel('Año')
plt.ylabel('Ingresos totales (millones de cop)')
plt.legend()
plt.grid(alpha=0.3)
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

2. Gráfico de envío promedio vs popularidad promedio

```python
envio_vs_popularidad = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Promedio_envio': [
        promedio_envio,
        promedio_envio2,
        promedio_envio3,
        promedio_envio4
        ],
    'Promedio_calificacion': [
        promedio_calificacion,
        promedio_calificacion2,
        promedio_calificacion3,
        promedio_calificacion4
        ]
})
envio_vs_popularidad
```

```python
plt.figure(figsize=(10,8))

plt.scatter(
    envio_vs_popularidad['Promedio_envio'],
    envio_vs_popularidad['Promedio_calificacion'],
    s=100
)

for i, row in envio_vs_popularidad.iterrows():
  offset = (10, 5)

  if row['Tienda'] == 'Tienda 1':
        offset = (-40, 10)

  plt.annotate(
      row['Tienda'],
      (row['Promedio_envio'], row['Promedio_calificacion']),
      textcoords="offset points",
      xytext=offset,
      ha='left'
  )

plt.title("Relación entre costo promedio de envío y calificación promedio")
plt.xlabel("Costo promedio de envío (COP)")
plt.ylabel("Calificación promedio")
plt.grid(alpha=0.3)
plt.show()
```
